```python
import pandas as pd
```

```python
dados1_2010 = pd.read_csv('/content/drive/MyDrive/7 days Code/emprestimos-20101.csv')
dados2_2010 = pd.read_csv('/content/drive/MyDrive/7 days Code/emprestimos-20102.csv')
dados1_2011 = pd.read_csv('/content/drive/MyDrive/7 days Code/emprestimos-20111.csv')
dados2_2011 = pd.read_csv('/content/drive/MyDrive/7 days Code/emprestimos-20112.csv')
dados1_2012 = pd.read_csv('/content/drive/MyDrive/7 days Code/emprestimos-20121.csv')
dados2_2012 = pd.read_csv('/content/drive/MyDrive/7 days Code/emprestimos-20122.csv')
dados1_2013 = pd.read_csv('/content/drive/MyDrive/7 days Code/emprestimos-20131.csv')
dados2_2013 = pd.read_csv('/content/drive/MyDrive/7 days Code/emprestimos-20132.csv')
```

# Importante Dados de Emprestimos


```python
emprestimos = pd.concat([dados1_2010, dados2_2010, dados1_2011, dados2_2011, dados1_2013, dados2_2013],ignore_index=True)
```

```python
emprestimos.shape
```

```python
seg_emprestimos = emprestimos.copy()
```

```python
seg_emprestimos
```

# Observando Frequência e excluindo duplicações

```python
seg_emprestimos.value_counts()
```

```python
seg_emprestimos =  seg_emprestimos.drop_duplicates()
```

```python
seg_emprestimos.value_counts()
```



# Importando dados de exemplares
